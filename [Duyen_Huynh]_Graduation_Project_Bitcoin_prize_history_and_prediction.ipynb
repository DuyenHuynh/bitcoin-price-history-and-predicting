#**2. Literature Review**

##**2.1. History of Bitcoin**

Bitcoin, created in 2009 by an anonymous individual (or group) known as Satoshi Nakamoto, is the first decentralized digital currency and is based on blockchain technology. Over the years, it has grown from an obscure digital experiment to a mainstream financial asset, impacting the global financial system. Here is an overview of Bitcoin's history, with some key milestones:

**1. The Birth of Bitcoin 👼 (2008-2009)**

- White Paper Release (2008): On October 31, 2008, Satoshi Nakamoto released a white paper titled "Bitcoin: A Peer-to-Peer Electronic Cash System" that described the concept of a decentralized, cryptographic digital currency.

- Genesis Block (2009): On January 3, 2009, the first block (Genesis Block) of the Bitcoin blockchain was mined. The block contained a message: "The Times 03/Jan/2009 Chancellor on brink of second bailout for banks," which highlighted the distrust in the traditional financial system.

- First Bitcoin Transaction (2009): Satoshi Nakamoto sent 10 BTC to developer Hal Finney, marking the first-ever Bitcoin transaction.

**2. Early Adoption and Development 📈 (2010-2013)**
- Bitcoin Pizza Day (2010): On May 22, 2010, programmer Laszlo Hanyecz made the first documented purchase using Bitcoin by paying 10,000 BTC for two pizzas. Today, this day is celebrated as "Bitcoin Pizza Day," highlighting how much Bitcoin’s value has increased.
- Mt. Gox Exchange Launch (2010): The first major Bitcoin exchange, Mt. Gox, was launched, becoming a popular platform for trading Bitcoin. However, it would later face a significant downfall.
Bitcoin Reaches Parity with the U.S. Dollar (2011): In February 2011, the price of 1 BTC reached $1 for the first time, marking a significant milestone in Bitcoin's history.
- The Silk Road (2011-2013): An online marketplace known for selling illegal goods used Bitcoin for transactions, drawing attention to its anonymity features and raising concerns about its use in illegal activities.

**3. The Rise, Fall, and Regulation 🍂(2013-2016)**
- Mt. Gox Collapse (2014): In early 2014, Mt. Gox, which was handling approximately 70% of Bitcoin transactions worldwide, declared bankruptcy after losing around 850,000 BTC, causing a significant drop in Bitcoin’s price and raising concerns about the security of digital currencies.
- Adoption by Mainstream Companies (2014): Companies like Microsoft, Overstock, and Dell started accepting Bitcoin as a form of payment, signaling growing acceptance in mainstream commerce.
- Blockchain Technology Gains Attention (2015): While Bitcoin itself was volatile, the underlying blockchain technology began to be recognized for its potential applications beyond cryptocurrency, such as in supply chain management and financial services.
- Segregated Witness (SegWit) Activation (2017): A protocol upgrade called Segregated Witness was implemented to address Bitcoin’s scalability issue by reducing the data size of each transaction.

**4. The Bull Run and Institutional Interest ❄️ (2017-2021)**
- Bitcoin Reaches 20,000 dollars (2017): Bitcoin's price surged to nearly 20,000 dollars in December 2017 due to increased demand and hype, largely fueled by retail investors.
- Introduction of Bitcoin Futures (2017): The launch of Bitcoin futures by CME and CBOE allowed institutional investors to engage with Bitcoin, bringing a new level of legitimacy to the asset.
- Price Correction (2018): Following the 2017 rally, Bitcoin’s price crashed by approximately 80%, leading to a prolonged "crypto winter."
- Institutional Adoption Accelerates (2020): Companies like MicroStrategy, Tesla, and Square began purchasing Bitcoin as part of their treasury management strategies, signaling a shift in institutional acceptance.
- Bitcoin Hits 60,000+ dollars (2021): Bitcoin’s price reached an all-time high of over 60,000 dollars, driven by institutional investment, the rise of decentralized finance (DeFi), and increased mainstream acceptance.

**5. Recent Developments 🔥 (2022-Present)**
- Regulatory Scrutiny (2022): Governments worldwide have increased their focus on regulating cryptocurrencies. The U.S., China, and the EU have all introduced various regulations aimed at increasing oversight of digital assets.
- Market Volatility and the Crypto Winter (2022): Following a period of significant growth, the cryptocurrency market experienced a downturn, with Bitcoin's price dropping significantly amid economic uncertainties, interest rate hikes, and the collapse of major crypto firms (e.g., FTX).
- The Rise of Bitcoin ETFs (2023): Bitcoin Exchange-Traded Funds (ETFs) have gained attention, with several proposals being reviewed by regulators. This development aims to make Bitcoin investments more accessible to traditional investors.
- Ongoing Institutional Interest and Technological Developments: Despite volatility, Bitcoin continues to see support from institutional investors and improvements in the network's scalability and sustainability, such as the Taproot upgrade that enhances privacy and smart contract capabilities.

Bitcoin’s history is marked by dramatic price swings, evolving technology, and shifts in public perception, making it a fascinating case study for understanding the intersection of finance, technology, and regulation.

##**2.3. Factors Influencing Bitcoin's Price**

Bitcoin's price is influenced by a combination of factors, including market dynamics, regulatory developments, and macroeconomic conditions. Here are the key factors with supporting data:

**1. Supply and Demand**
- Limited Supply: Bitcoin has a capped supply of 21 million coins, creating scarcity. As of 2024, around 19.5 million BTC have been mined. This scarcity can drive prices up as demand increases.
- Halving Events: Bitcoin undergoes a "halving" roughly every four years, reducing the reward miners receive for validating transactions by half. This has historically led to price increases due to reduced supply flow.

  For example:
  - 2012 Halving: Bitcoin's price surged from 12 dollars to over 1,000 dollars within a year.
  - 2016 Halving: Price increased from 650 dollars to nearly 20,000 dollars by the end of 2017.
  - 2020 Halving: Bitcoin rose from around 9,000 dollars to over 60,000 dollars by early 2021.

**2. Market Sentiment**

- Investor Hype and Media Attention: Bitcoin's price is sensitive to news and social media activity. Positive news, such as institutional adoption, often leads to price surges, while negative news, like exchange hacks or regulatory crackdowns, can cause sharp declines.
- Google Trends and Social Media Sentiment: Studies have found a correlation between Bitcoin-related search trends and price movements. For example, a 2015 study showed that spikes in Google search activity for "Bitcoin" often coincided with price increases.

**3. Regulatory Developments**

- Government Actions and Regulations: Regulations and legal actions by governments can heavily influence Bitcoin’s price.

  For example:
  - China's Crackdowns: China has implemented multiple bans on Bitcoin-related activities, including a 2021 crackdown on mining and trading, which caused Bitcoin's price to drop by around 30% within weeks.
  - U.S. Regulations: Positive news about potential approval of Bitcoin Exchange-Traded Funds (ETFs) in the U.S. has led to price increases, as it signals broader acceptance.

- Tax and Legal Frameworks: Countries that provide a legal framework for cryptocurrency use, such as Japan and Switzerland, tend to boost Bitcoin adoption and price stability.

**4. Macroeconomic Factors**

- Inflation and Currency Devaluation: Bitcoin is often seen as a hedge against inflation, attracting investors during times of economic uncertainty. For example, during the COVID-19 pandemic and subsequent economic stimulus efforts, Bitcoin's price surged from around 7,000 dollars in March 2020 to over 60,000 dollars in 2021.

- Interest Rates and Monetary Policy: Low interest rates and quantitative easing can increase demand for alternative assets like Bitcoin. Conversely, rising interest rates may lead to decreased interest in riskier assets, including cryptocurrencies.

**5. Technological Developments and Network Upgrades**
- Network Upgrades (e.g., SegWit, Taproot): Improvements to Bitcoin's blockchain, such as the activation of Segregated Witness (SegWit) in 2017 and Taproot in 2021, which enhanced transaction speed and privacy, have led to positive market responses.
- Development of the Lightning Network: This second-layer solution aims to improve transaction speeds and lower fees, making Bitcoin more appealing for everyday transactions, which can positively impact its price.

**6. Adoption by Institutional Investors**
- Institutional Investments: The entry of institutional players like MicroStrategy, Tesla, and Grayscale Investments has brought significant capital into the market.

  For example:
  - MicroStrategy’s Bitcoin Purchases: The company's ongoing investment in Bitcoin, starting in 2020, helped drive a price increase from around 10,000 dollars to over 50,000 dollars in early 2021.
  - Bitcoin Exchange-Traded Funds (ETFs): The launch of Bitcoin futures ETFs in 2021 in the U.S. provided more access to institutional investors, contributing to price appreciation.

**7. Market Liquidity and Trading Volume**
- Exchange Liquidity: More liquid markets can stabilize prices, while less liquid markets may exhibit greater volatility. Sudden large trades in illiquid markets can cause price swings.
- Trading Volume Correlation: There is a strong correlation between trading volume and price movements. Periods of high trading volume often coincide with significant price changes.

**8. Geopolitical Events**
- Global Uncertainty and Political Instability: Events such as the war in Ukraine or U.S.-China trade tensions can drive investors to Bitcoin as a safe-haven asset.
- Sanctions and Capital Controls: When countries implement capital controls or face economic sanctions, Bitcoin adoption often increases as an alternative way to move and store value.

**9. Competing Cryptocurrencies and Blockchain Projects**
- Altcoins and "Flippening" Speculation: The emergence of other cryptocurrencies like Ethereum and their market developments can affect Bitcoin's dominance. For instance, the rise of decentralized finance (DeFi) and non-fungible tokens (NFTs) on the Ethereum blockchain has sometimes led to shifts in investment from Bitcoin to other digital assets.
- Interoperability and Cross-Chain Solutions: Projects enabling cross-chain compatibility (e.g., Polkadot, Cosmos) can impact the flow of capital between Bitcoin and other cryptocurrencies.

###**Fixing Data Types**

```python
def convert_to_float(value):
    #Convert a volume string with suffixes (K, M, B) into its numeric value.

    #Parameters:
    #value (str): The input volume string with a possible suffix.

    #Returns:
    #float: The numeric value corresponding to the input volume string,
           #or 0 if the input is invalid.
    # Trim any whitespace
    value = value.strip()
    multiplier = 1

    # Check for suffix and set multiplier
    if 'K' in value:
        value = value.replace("K", "")
        multiplier = 1000
    elif 'M' in value:
        value = value.replace("M", "")
        multiplier = 1000000
    elif 'B' in value:
        value = value.replace("B", "")
        multiplier = 1000000000

    # Check for invalid input (like negative sign)
    if "-" in value:
        return 0

    try:
        # Convert to float and apply multiplier
        numeric_value = float(value) * multiplier
        return numeric_value
    except ValueError:
        # Handle invalid numeric input
        return 0
```

```python
price_data['Volume'] = price_data['Volume'].apply(lambda x: convert_to_float(x))
```

###**Time Series Analysis**
Time Series data consists of observations collected at various points in time and has key characteristics such as trend, seasonality, and noise. Forecasting involves making predictions about the future based on historical and current data. This project aims to conduct a Time Series Analysis on historical Bitcoin price data.

As seen in the Data Exploration section, Bitcoin prices have shown high volatility and inconsistency over the years, making Time Series analysis challenging. Here we try to explore the different Time series forecasting model, which will be discussed in the following sections.

**Stationarity Check: Close Price**

Let's view if the time series of Bitcoin close price is stationary.

```python
def stationarity_check(ts):
    plt.plot(ts, label='Original')
    plt.plot(ts.rolling(window=30, center=False).mean(), label='Rolling Mean')
    plt.plot(ts.rolling(window=30, center=False).std(), label='Rolling Std')
    plt.grid()
    plt.legend()
    plt.show()

    adf = adfuller(ts, autolag='AIC')
    padf = pd.Series(adf[:4], index=['T Statistic','P-Value','#Lags Used','#Observations Used'])
    for k,v in adf[4].items():
        padf['Critical value {}'.format(k)]=v
    print(padf)

stationarity_check(price_data['Close'])
```

Interpreting Results:
- If the p-value is below 0.05, this suggests that the series is stationary (reject the null hypothesis).
- If the test statistic is less than the critical values at 1%, 5%, or 10%, it means the time series is stationary at the corresponding confidence level.

Since the p-value is higher than 0.05 and  the test statistic is higher than the 10% critical value, the null hypothesis cannot be rejected. This implies that the time series of the Bitcoin price is non-staionary.

**Daily Trends: Auto-Correlation and Partial Auto-Correlation**

Consider adding `ACF` and `PACF` plots to visualize autocorrelation and identify if further differencing is needed.

- `Autocorrelation Function (ACF)`: The plot summarizes the correlation of an observation with lag values. The x-axis shows the lag and the y-axis shows the correlation coefficient between -1 and 1 for negative and positive correlation.
- `Partial Autocorrelation Function (PACF)`: The plot summarizes the correlations for an observation with lag values that is not accounted for by prior lagged observations. We can get a basic picture of the parameter interval, and using this heuristic we can with help decide which are the best p,q,d for ARIMA.

```python
# ACF Plot
gp = plot_acf(price_data['Close'])
plt.show()
```

Interpreting the ACF Plot:
- Significant spikes outside the confidence interval bands (usually ±1.96/√n) indicate significant autocorrelation at specific lags.
- Regularly spaced spikes suggest seasonality in the data.
- Slowly decaying ACF suggests the presence of a trend.

The results do make sense, as we expect today's price close will indeed affect future prices, however we do not see any clear cycles in the time series.

Therefore, we may want to look at the auto-correlations deeper, using Partial Auto-Correlation (PAC). The PACF helps to identify the direct relationship between observations at different lags, excluding the influence of intermediate lags. In a time series essentially it describes the correlation between points t and t+n conditional on the time series values between them.

```python
# PACF Plot
gp = plot_pacf(price_data['Close'], method='ywm')
plt.show()
```

Interpreting the PACF Plot:
- Significant lags: A significant spike at lag k indicates that the kth lag has a strong direct effect on the current value, after accounting for the effects of lags 1 through k-1.
- Cut-off after a few lags: If the PACF cuts off after a few lags, it suggests the order of the AR term in ARIMA should match this number.

We can see from the above diagram that only the 1st daily lagged value have a clear statistical significance with regards to their partial auto-correlations. The PACF cuts off sharply while the ACF has a gradual decline, which indicates an autoregressive (AR) process.

**Decomposing Time Series: trend, seasonality, and residuals (daily)**

Let's decompose the Bitcoin prices (close) time series into trend, seasonal, and residual components to better understand the non-stationarity.
- Trend
- Seasonality
- Residual (in the classical time series paradigm, it is attributed to be purely random or irregular; such fluctuations are considered to be unforeseen, uncontrollable, unpredictable, and are erratic within the classical time series analysis)

```python
result_add = sm.tsa.seasonal_decompose(x=price_data['Close'], model='additive', extrapolate_trend='freq', period=365)
result_add.trend

frame = {
    'price_date': price_data['Date'],
    'observed': result_add.observed,
    'trend': result_add.trend,
    'seasonal': result_add.seasonal,
    'residuals': result_add.resid
}

# Creating DataFrame by passing Dictionary
decomposed_ts_df = pd.DataFrame(frame).reset_index()
```

```python
fig = go.Figure([
    go.Scatter(
        name='observed',
        x=decomposed_ts_df['price_date'],
        y=decomposed_ts_df['observed'],
        mode='lines',
        marker=dict(color=px.colors.qualitative.Prism[1], size=3),
        showlegend=True
    ),
    go.Scatter(
        name='trend',
        x=decomposed_ts_df['price_date'],
        y=decomposed_ts_df['trend'],
        mode='lines',
        marker=dict(color='red', size=3),
        showlegend=True
    ),
    go.Scatter(
        name='seasonal',
        x=decomposed_ts_df['price_date'],
        y=decomposed_ts_df['seasonal'],
        mode='lines',
        marker=dict(color=px.colors.qualitative.Prism[0]),
        line=dict(width=2),
        showlegend=True
    ),
    go.Scatter(
        name='residual',
        x=decomposed_ts_df['price_date'],
        y=decomposed_ts_df['residuals'],
        mode='markers',
        marker=dict(color=px.colors.qualitative.Prism[2], size=2),
        showlegend=True
    )
])
fig.update_layout(
    yaxis_title='price, USD',
    title='Decomposed Daily Time Series: Bitcoin Close Price',
    #hovermode="x"
)
fig.show()
```

**Keys:**

- The trend in Bitcoin prices is curvilinear (non-linear), with the stochastic trend being the primary factor influencing observed Bitcoin prices.
- While seasonal patterns exist on 3-, 6-, and 12-month scales, they contribute less to the observed values than the trend does.
- The residual (random) component, though smaller than the trend, sometimes exceeds the seasonal component in magnitude.
- The impact of the residual on observed values surpasses that of the seasonal component at local minima and maxima in the Bitcoin price curve.

Notes:

- Stochastic trends are related to unit roots, integration, and differencing.

##**3.3. Model Selection**

**Top 2 most popular models**

Predicting Bitcoin prices is a complex task, and several machine learning models are commonly used for this purpose. Here are the top 2 most popular models for Bitcoin price prediction, along with their advantages and disadvantages:

1. `Autoregressive Integrated Moving Average (ARIMA)`

  ARIMA is a popular statistical model used for time series forecasting. It models the data based on its own past values (autoregression), the differencing of the data to achieve stationarity (integrated), and lagged errors (moving average).

  - Advantages:
Simple and Interpretable: ARIMA models are relatively easy to implement and understand. The coefficients provide clear insights into the behavior of the data.
Effective for Short-Term Prediction: When Bitcoin prices exhibit a stable trend (at least in the short term), ARIMA can be very effective.
Minimal Data Requirements: Unlike deep learning models, ARIMA doesn’t require vast amounts of data to train.
  - Disadvantages:
Struggles with Non-Stationary Data: Bitcoin price data is often highly non-stationary, and ARIMA requires the data to be stationary for reliable predictions, which can be a major limitation.
Limited to Linear Relationships: ARIMA assumes linear relationships between past values and future values, which may not capture the complex, nonlinear nature of Bitcoin price movements.
Not Suitable for Long-Term Predictions: ARIMA works best for short-term forecasting and struggles with the long-term volatility of Bitcoin prices.

2. `Long Short-Term Memory (LSTM)` - A type of Recurrent Neural Network (RNN)

  LSTMs are widely used for time series forecasting, including Bitcoin price prediction, due to their ability to capture long-term dependencies in sequential data. They are good at modeling temporal relationships. LSTMs use a cell state to store information about past inputs. This cell state is updated at each step of the network, and the network uses it to make predictions about the current input. The cell state is updated using a series of gates that control how much information is allowed to flow into and out of the cell.

  - Advantages:
Captures Long-Term Dependencies: LSTMs can retain and utilize information from previous time steps, making them effective for time series predictions where past values influence future ones.
Handles Non-Stationary Data: Bitcoin prices are volatile and non-stationary, and LSTMs can adjust to these fluctuations better than simpler models.
Adaptable to Complex Data: LSTMs can handle multidimensional input, which allows the inclusion of additional features like market sentiment or transaction volumes.

  - Disadvantages:
Requires Large Data: LSTMs need a large amount of training data to achieve high accuracy, which can be a problem if only short-term data is available.
Complexity: LSTMs are computationally expensive, slower to train, and require tuning several hyperparameters (e.g., number of layers, neurons, dropout rate).
Overfitting: If not tuned properly, LSTMs can overfit the training data, leading to poor generalization to new data.


Summary of Models:

| **Model**         | **Advantages**                                         | **Disadvantages**                                    |
|-------------------|-------------------------------------------------------|------------------------------------------------------|
| **ARIMA**         | Simple, interpretable, good for short-term forecasting, minimal data requirements | Struggles with non-stationary data, only models linear relationships, not suitable for long-term |
| **LSTM**          | Captures long-term dependencies, handles non-stationary data, adapts to complex data | Requires large data, complex, risk of overfitting     |


Each of these models has its strengths and limitations depending on the context of prediction (short-term vs. long-term, available data, etc.). For Bitcoin price prediction, LSTM tends to perform better in volatile, time-dependent situations, while ARIMA can be effective for short-term linear trends. ##**4.1. ARIMA**

Let’s take a look at ARIMA, which is one of the most popular (if not the most popular) time series forecasting techniques. In order to really understand ARIMA, we need to deconstruct its building blocks. Once we have the components down, it will become easier to understand how this time series forecasting method works as a whole.
- `Autoregressive (AR) part`
The Autoregressive (AR) component builds a trend from past values in the AR framework for predictive models. For clarification, the 'autoregression framework' works like a regression model where you use the lags of the time series' own past values as the regressors.

- `Integrated (I) part`
The Integrated (I) part involves the differencing of the time series component keeping in mind that our time series should be stationary, which really means that the mean and variance should remain constant over a period of time. Basically, we subtract one observation from another so that trends and seasonality are eliminated. By performing differencing we get stationarity. This step is necessary because it helps the model fit the data and not the noise.

- `Moving average (MA) part`
The moving average (MA) component focuses on the relationship between an observation and a residual error. Looking at how the present observation is related to those of the past errors, we can then infer some helpful information about any possible trend in our data.

We can consider the residuals among one of these errors, and the moving average model concept estimates or considers their impact on our latest observation. This is particularly useful for tracking and trapping short-term changes in the data or random shocks. In the (MA) part of a time series, we can gain valuable information about its behavior which in turn allows us to forecast and predict with greater accuracy.

Now let's try to build ARIMA model using our data.

The p-value < 0.05, so data is stationary.

**Determining ARIMA Parameters (p, d, q)**

- `p`: The number of lag observations (autoregressive term), also called the lag order and determined from the PACF plot by the point where the PACF cuts off sharply.
- `d`: The number of times the data is differenced, also called the degree of differencing and set to the number of times differencing was applied (typically 1 if the series was non-stationary).
- `q`: The size of the moving average window, also called the degree of differencing and determined from the ACF plot by the point where the ACF cuts off sharply.


```python
# Plot ACF and PACF
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
plot_acf(train_data_arima_diff, lags=20, ax=axes[0])
axes[0].set_title('ACF Plot')
plot_pacf(train_data_arima_diff, lags=20, ax=axes[1])
axes[1].set_title('PACF Plot')
plt.show()
```

Based on the ACF and PACF plots:

- The ACF plot drops off after lag 1, q may be 1.
- The PACF plot cuts off after lag 1, p may be 1.

**Build and Train the ARIMA Model**

```python
#2. Build model
model = ARIMA(train_data_arima, order=(1, 1, 1))
model_fit = model.fit()
print(model_fit.summary())
```

##**4.2. LSTM**

Long Short Term Memory (LSTMs) is capable of learning long-term dependencies. LSTMs are explicitly designed to avoid the long-term dependency problem. Remembering information for long periods of time is practically their default behavior, not something they struggle to learn!

The key to LSTMs is the cell state, the horizontal line running through the top of the diagram.

The cell state is kind of like a conveyor belt. It runs straight down the entire chain, with only some minor linear interactions. It’s very easy for information to just flow along it unchanged.

The LSTM does have the ability to remove or add information to the cell state, carefully regulated by structures called gates.

Gates are a way to optionally let information through. They are composed out of a sigmoid neural net layer and a pointwise multiplication operation. The sigmoid layer outputs numbers between zero and one, describing how much of each component should be let through. A value of zero means “let nothing through,” while a value of one means “let everything through!”

An LSTM has three of these gates, to protect and control the cell state.

**Split the Data**

Split the data into training, validation and out-of-time test sets. We have 10 years of daily data:

- Use the first 7 years for training.
- Use the next 2 years for validation.
- Reserve the final year for out-of-time testing.

```python
# 1. Split data into train, validation, and out-of-time test sets
train_data = close_price[:int(len(close_price) * 0.7)]
val_data = close_price[int(len(close_price) * 0.7):int(len(close_price) * 0.9)]
out_of_time_test_data = close_price[int(len(close_price) * 0.9):]
```

**Normalize the Data**

Normalization is a technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information.

LSTMs perform better when the data is scaled. Use MinMaxScaler from scikit-learn to scale the price between 0 and 1. For each value in a feature, MinMaxScaler subtracts the minimum value in the feature and then divides by the range. The range is the difference between the original maximum and original minimum. MinMaxScaler preserves the shape of the original distribution.


```python
# 2. Scale data
# Select only the numerical features for scaling
scaler = MinMaxScaler()
train_data_scaled = scaler.fit_transform(train_data.values.reshape(-1, 1))
val_data_scaled = scaler.transform(val_data.values.reshape(-1, 1))
out_of_time_test_scaled = scaler.transform(out_of_time_test_data.values.reshape(-1, 1))
```

**Create Sequences**

LSTM models work on sequences of data, so create a sliding window of sequences.

```python
# 3. Create sequences
# Function to create sequences
def create_sequences(data, sequence_length):
    X = []
    y = []

    for i in range(len(data) - sequence_length - 1):
        X.append(data[i:(i + sequence_length), 0])  # Extract values for consistency in numpy array
        y.append(data[i + sequence_length, 0])      # The next value following the sequence

    return np.array(X), np.array(y)

# Sequence length
sequence_length = 50

# Generate sequences for each dataset
X_train, y_train = create_sequences(train_data_scaled, sequence_length)
X_val, y_val = create_sequences(val_data_scaled, sequence_length)
X_test, y_test = create_sequences(out_of_time_test_scaled, sequence_length)

# Reshape the sequences for LSTM input (samples, time steps, features)
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Print shapes to confirm
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

**Build the LSTM Model**

Now that the data is preprocessed, we can define the LSTM model. LSTM models have three key layers:

- LSTM Layers: To capture the time dependencies in the data.
- Dropout Layers: To prevent overfitting by randomly setting some weights to zero during training.
- Dense Layer: A fully connected layer to output the final prediction.

```python
# 4. Define the LSTM model
model_input = layers.Input(shape = X_train.shape[1:])
model_lstm = layers.LSTM(15)(model_input)
model_dense = layers.Dense(1)(model_lstm)
model = models.Model(inputs=model_input, outputs=model_dense)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()
```

**Train the Model**

Train the model using the training data for a specified number of epochs.

```python
# 5. Train the model
# Define early stopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model with early stopping
history = model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Plot training history
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

**Make Predictions**

After training the model, use it to predict the future Bitcoin prices. Ensure that the scaling inverted to get back to the original price values.

```python
# 6. Make predictions on the out-of-time test set
predictions = model.predict(X_test).reshape(-1, 1)  # Reshape to (n_samples, 1)

# Inverse transform predictions and y_test
predicted_price = scaler.inverse_transform(predictions)
actual_price = scaler.inverse_transform(y_test.reshape(-1, 1))

# 7. Visualize the actual vs predicted prices
plt.plot(actual_price, label='Actual Price')  # Actual prices from test period
plt.plot(predicted_price, label='Predicted Price')  # Predicted prices
plt.title('Bitcoin Price Prediction')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

**Evaluate the Model**

Evaluate the model using metrics like Mean Squared Error (MSE) or Root Mean Squared Error (RMSE) to understand how well the model performed.

```python
# 8. Evaluate model performance on the out-of-time test set
#Calculate MSE, MAE, RMSE
mse = mean_squared_error(actual_price, predicted_price)
mae = mean_absolute_error(actual_price, predicted_price)
rmse = math.sqrt(mse)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'RMSE: {rmse}')

#Calculate MAPE
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mape = mean_absolute_percentage_error(actual_price, predicted_price)
print(f'MAPE: {mape}')
```

- RMSE (2,318.03) suggests the model’s error magnitude, with larger deviations from the actual prices more heavily penalized.
- A MAPE of 3.04% indicates that, on average, the model's predictions are off by just 3.04% of the actual Bitcoin prices. A lower MAPE, generally under 10%, is often seen as an indicator of high forecasting accuracy in time series models.

These metrics suggest that the LSTM model performs quite well, especially with a low MAPE (3.04%) indicating accurate predictions relative to actual prices.
